```python
import numpy as np
import matplotlib.pyplot as plt

# Lattice constant
a = 1  # You can adjust this value as needed
# Size of the Brillouin zone
s = 4 * np.pi / (3 * a)
# Angles for hexagon vertices
theta = np.linspace(0, 2 * np.pi, 7)[:-1]
# Hexagon vertices
x = s * np.cos(theta)
y = s * np.sin(theta)

fig, ax = plt.subplots()

# Plot hexagon
ax.plot(np.append(x, x[0]), np.append(y, y[0]), 'k-')

k_path = np.array([[0, 0], [2/3, 1/3], [.5, .5], [1/3, 2/3], [0, 0], [.5, .5]])
label = [r'$\Gamma $',r'$K$', r'$M$', r'$K^\prime$', r'$\Gamma $', r'$M$']
k_pts = k_path @ recip_vecs

for i in range(4):
    ax.scatter(k_pts[i,0], k_pts[i,1], zorder=2, c='k')
    ax.text(k_pts[i,0]+0.4, k_pts[i,1], label[i], fontsize=12, ha='center', va='center')

# Plot settings
ax.axis('equal')
# plt.xlabel('$k_x$')
# plt.ylabel('$k_y$')
# plt.title('Graphene Brillouin Zone')
# plt.grid(True)
ax.set_xticks([])
ax.set_yticks([])
ax.set_axis_off()
```
